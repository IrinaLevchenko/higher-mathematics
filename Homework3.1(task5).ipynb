```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2 * X + 5 * Y
Z1 = 2 * X + 5 * Y + 20

ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z1)
plt.show()
```
