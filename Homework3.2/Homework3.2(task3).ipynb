```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polar_to_decart(R, alfa):
    x = R * np.cos(alfa)
    y = R * np.sin(alfa)
    return (x, y)

print(polar_to_decart(10, 30))
```

```python
phi = np.arange(0., 2., 1./180.)*np.pi 

plt.polar(phi, [1.4] * len(phi))

plt.show()
```
