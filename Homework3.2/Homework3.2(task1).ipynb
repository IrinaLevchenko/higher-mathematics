```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2 * np.pi, 3 * np.pi, 201)

plt.plot(x, np.cos(x), 'r')
plt.plot(x, np.cos(x - 2), 'y')
plt.plot(x, np.cos(x) + 2, 'g')
plt.plot(x, 2 * np.cos(x), 'b')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True) 
```
