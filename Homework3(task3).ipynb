```python
%matplotlib inline
import matplotlib.pyplot as plt
import cmath
```

```python
xO = []
x = []
yO = []
y = []

for i in range(100000):
    R = 100000
    xO.append(-i)
    x.append(i)
    yO.append(cmath.sqrt(R ** 2 - i ** 2))
    y.append(-cmath.sqrt(R ** 2 - i ** 2))

plt.figure(figsize = (5, 5))
plt.title('График окружности')
plt.plot(xO, yO, color='blue')
plt.plot(x, y, color='blue')
plt.plot(xO, y, color='blue')
plt.plot(x, yO, color='blue')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cmath

xO = []
x = []
yO = []
y = []

for i in range(40000):
    a = 40000
    b = 20000   
    xO.append(-i)
    x.append(i)
    yO.append(cmath.sqrt(b ** 2 - b ** 2 * (i ** 2 / a ** 2)))
    y.append(-cmath.sqrt(b ** 2 - b ** 2 * (i ** 2 / a ** 2)))
    
plt.title('График эллипса')
plt.plot(xO, yO, color='red')
plt.plot(x, y, color='red')
plt.plot(xO, y, color='red')
plt.plot(x, yO, color='red')

    
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cmath

xO = []
x = []
yO = []
y = []

for i in range(100):
    a = 60
    b = 40
    yO.append(-i)
    y.append(i)
    xO.append(cmath.sqrt(a ** 2 + a ** 2 * (i ** 2 / b ** 2)))
    x.append(-cmath.sqrt(a ** 2 + a ** 2 * (i ** 2 / b ** 2)))
  
plt.title('График гиперболы')
plt.plot(xO, yO, color='y')
plt.plot(x, y, color='y')
plt.plot(xO, y, color='y')
plt.plot(x, yO, color='y')
```
